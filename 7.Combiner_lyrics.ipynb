```python
from google.colab import drive
import pandas as pd
import numpy as np
from tqdm import tqdm as tq
import glob

drive.mount('/content/drive')
```

```python
# 경로 안의 .xlsx 형식의 파일 목록을 가져와서 list화
file_format = ".xlsx"
file_path = "/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/lyrics"
file_list = glob.glob(f"{file_path}/*{file_format}")

# 병합 결과를 담을 pandas DataFrame 객체 생성
merge_df = pd.DataFrame()

for i in tq(file_list):
# file_list에 담겨있는 파일 경로 및 파일명을 통해 엑셀파일 읽어오기
  file_df = pd.read_excel(i)
# 임시로 사용할 변수 temp_df에 불러온 엑셀파일의 columns 중 작성한 list의 column만을 DataFrame 객체로 저장
  temp_df = pd.DataFrame(file_df)
# 병합결과를 담을 객체에 temp_df의 객체를 삽입
# ignore_index는 기존 파일의 index를 무시하는 옵션
  merge_df = merge_df.append(temp_df, ignore_index=True)

# 병합 결과를 담은 pandas DataFrame 객체를 csv파일로 저장(excel로 처리하면 데이터가 너무 많아서 오류 발생)
merge_df.to_csv("/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/Processing/lyrics_all.csv", index=False, encoding='utf-8')

print("############ 파일 병합이 모두 완료되었습니다. ############")
print("############ 추출된 lyrics의 총 갯수는 "+str(len(merge_df)) +"개 입니다. ############")
```
