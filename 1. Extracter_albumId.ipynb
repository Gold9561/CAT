3. json 파일을 열고 data변수에 로드

```python
import json
import pandas as pd

with open('/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/song_meta.json',encoding='UTF-8') as songFile:
    data = json.load(songFile)
```

4. data 변수에서 'album_id'를 key로 가진 값을 추출하여 리스트 형식의 albumId에 추가, num은 no에 추가

```python
num=0
no = []
albumId = []
for i in data:
    # print(num)
    albumId.append(i['album_id'])
    # print(albumId[num])
    num+=1
    no.append(num)
```

5. pandas를 사용하여 no와 albumId를 index로 가진 DataFrame 생성 및 엑셀파일 생성, drop_duplicate 명령을 이용하여 중복값 제거후 중복값 제거된 파일 생성

```python
df=pd.DataFrame({"no.":no,"album_id":albumId})

df.to_excel("/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/앨범Id_test.xlsx", encoding='utf-8', index=False)

df2=df.drop_duplicates(['album_id'], keep='first')

df2.to_excel("/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/앨범Id_중복값제거_test.xlsx", encoding='utf-8', index=False)
```
