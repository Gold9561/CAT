2. glob을 통해 폴더 내의 파일 개수를 확인하여 모든 곡id의 추출이 완료되었는지 확인

```python
# 경로 안의 .xlsx 형식의 파일 목록을 가져와서 list화
file_format = ".xlsx"
file_path = "/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/songId"
file_list = glob.glob(f"{file_path}/*{file_format}")

# 파일의 개수를 카운팅
file_num=0
for i in file_list:
  file_num = file_num+1


# 1,437 개의 파일이 존재하는지 검사
if file_num == 1437:
  print("############ "+str(file_num)+" 개의 파일이 존재합니다. ############")
  print("############ song_id의 추출이 모두 완료되었습니다. ############")
else:
  print("############ song_id의 추출이 완료되지 않았습니다. ############")

```

3. 추출된 곡id를 모두 병합하여 한개의 엑셀파일로 작성

```python
# 병합 결과를 담을 pandas DataFrame 객체 생성
merge_df = pd.DataFrame()

# 병합할 column명 list 작성
columns = ["song_id"]

for i in file_list:
# file_list에 담겨있는 파일 경로 및 파일명을 통해 엑셀파일 읽어오기
  file_df = pd.read_excel(i)
# 임시로 사용할 변수 temp_df에 불러온 엑셀파일의 columns 중 작성한 list의 column만을 DataFrame 객체로 저장
  temp_df = pd.DataFrame(file_df, columns=columns)
# 병합결과를 담을 객체에 temp_df의 객체를 삽입
# ignore_index는 기존 파일의 index를 무시하는 옵션
  merge_df = merge_df.append(temp_df, ignore_index=True)

# 병합 결과를 담은 pandas DataFrame 객체를 csv파일로 저장(excel로 처리하면 데이터가 너무 많아서 오류 발생)
merge_df.to_csv("/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/songId_all.csv", index=False, encoding='utf-8')

print("############ 파일 병합이 모두 완료되었습니다. ############")
print("############ 추출된 song_id의 총 갯수는 "+str(len(merge_df)) +"개 입니다. ############")
```

4. 분할파일로 저장

```python
# 파일 분할 시 한 파일에 들어갈 song_id의 갯수를 지정
chunksize = 1000

num = 0

for chunk in np.array_split(merge_df, len(merge_df) // chunksize):
  chunk.to_excel('/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/chunk_songId/file_{:02d}.xlsx'.format(num), encoding='utf-8', index=False)
  num = num + 1

print("############ 파일 분할이 모두 완료되었습니다. ############")
print("############ 분할된 파일의 총 갯수는 "+str(num) +"개 입니다. ############")
```
