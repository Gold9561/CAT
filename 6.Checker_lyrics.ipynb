2. 가사 크롤링 진행 상황 확인

```python
fileNum = list(range(0,2429))
filelist = 0
nofile = []
nofilelist = 0

for x in tq(fileNum):
  if os.path.isfile('/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/lyrics/lyrics_{:02d}.xlsx'.format(x))==False:
    nofile.append(x)
    nofilelist = nofilelist + 1
  else:
    filelist = filelist + 1

progress_lyrics =  round((filelist/2429.0)*100,2)

print("\n\n##############   작성된 파일 수 : "+str(filelist)+ " 개 ##############")
print("############## 미작성된 파일 수 : "+'%4d' % nofilelist+ " 개 ##############")
print("############## 가사 수집 진행률 : " + str(progress_lyrics)+" % ##############\n")

npnofile=np.array(nofile)
npnofile
```
