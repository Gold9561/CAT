2. 각각의 모듈 import 및 구글드라이브 연결

```python
from google.colab import drive
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

drive.mount('/content/drive')

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=options)
```

3. 가사 크롤링 진행

```python
#fileNum 수정하며 진행
fileNum = list(range(47,50))

for x in fileNum:
  # songid가 들어있는 분할파일을 읽어옴
  df = pd.read_excel('/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/chunk_songId/file_{:02d}.xlsx'.format(x))
  df1 = df.astype(str)
  state = 0

  SONG_NAME=[]
  ARTIST=[]
  ALBUM=[]
  RELEASE_DATE=[]
  GENRE=[]
  LYRIC_FULL=[]

  for songId in df1['song_id']:
    state=state + 1
    driver.get("https://www.melon.com/song/detail.htm?songId="+songId)
    time.sleep(3)
    song_name=driver.find_element(by=By.CLASS_NAME, value="song_name")
    artist=driver.find_element(by=By.CLASS_NAME, value="artist")
    meta=driver.find_element(by=By.CLASS_NAME, value="meta")
    SONG_NAME.append(song_name.text)
    ARTIST.append(artist.text)
    meta_list = meta.text.split('\n')
    ALBUM.append(meta_list[1])
    RELEASE_DATE.append(meta_list[3])
    GENRE.append(meta_list[5])
    try:
        lyric_full=driver.find_element(by=By.CLASS_NAME, value="lyric")
        LYRIC_FULL.append(lyric_full.text)
    except:
        LYRIC_FULL.append("")
    if state == 250 or state == 500 or state == 750:
      print("현재 " + str(x) + "번 파일의 작업이 " + str(state/10) + "% 완료되었습니다.")

  df=pd.DataFrame({"Song_Name":SONG_NAME,"Artist":ARTIST,"Album":ALBUM, "Release_Date":RELEASE_DATE,"Genre":GENRE,"Lyric":LYRIC_FULL})
  df.to_excel('/content/drive/MyDrive/team cat/02.입력 데이터처리 출력 과정/Scraper/data/lyrics/lyrics_{:02d}.xlsx'.format(x), encoding='utf-8', index=False)

  print("\n######################## "+str(x)+"번째 파일 작성 완료 ########################\n")
```
