```python
data['Lyric'] = data['Lyric'].replace(np.nan,'')
lyr = data['Lyric']
lyr
```

```python
def clean_text(text):
    cleaned_text = re.sub('[a-zA-z]','',str(text))
    cleaned_text = re.sub('[0-9]','',cleaned_text)
    cleaned_text = re.sub('[\{\}\[�\]\/?.,;:|©\“”‘)\u2005*~’`!^\-_+<>@\#$%&\\\=\(\'\"\♥\♡\ㅋ\ㅠ\ㅜ\ㄱ\ㅎ\ㄲ\ㅡ\접기]','',cleaned_text)
    cleaned_text = re.sub('[\n]',' ',cleaned_text)
    return cleaned_text

lyr_kor = []
for i in tq(range(len(lyr))):
  lyr_kor.append(clean_text(lyr[i]))

data['lyr_kor']=lyr_kor 
data
```

```python
data=data[data['lyr_kor'].str.strip().astype(bool)]
data = data.reset_index(drop=True)
data
```

```python
def blank_del(text):
  blank_del_text = " ".join(str(text).split())
  blank_del_text = blank_del_text.replace(" ","")  
  return blank_del_text

lyr_kor=data['lyr_kor']
lyr_blank_del = []
for i in tq(range(len(lyr_kor))):
  lyr_blank_del.append(blank_del(lyr_kor[i]))

data['lyr_blank_del']=lyr_blank_del 
data
```

```python
data.drop(data[data['lyr_blank_del'].map(len) < 70].index, inplace=True)
data = data.reset_index(drop=True)
data
```
